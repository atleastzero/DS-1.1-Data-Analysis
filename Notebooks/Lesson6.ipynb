```python
def count_lower(ls, threshold):
    return len(ls[ls <= threshold])/len(ls)
```

```python
def cdf_list(ls, step_size):
    new_ls = []
    threshold_range = range(int(ls.min()), int(ls.max())+1, step_size)
    for i in threshold_range:
        new_ls.append(count_lower(ls, i))
    return new_ls
```

```python
import matplotlib.pyplot as plt

plt.plot(cdf_list(ti["Age"].dropna(), 1))
plt.grid()
```
